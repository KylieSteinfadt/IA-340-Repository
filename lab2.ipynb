## Group & Visualize the Data

```python
average_price_by_form = df.groupby('Form')['CupEquivalentPrice'].mean().reset_index()
average_price_by_form.head()
```

```python
import matplotlib.pyplot as plt
import io
```

```python
plt.bar(average_price_by_form['Form'], average_price_by_form['CupEquivalentPrice'], color=['skyblue', 'lightgreen', 'salmon', 'yellow'])
plt.title('Average Price per Cup Equivalent by Vegetable Type')
plt.xlabel('Vegetable Type')
plt.ylabel('Average Price per Cup ($)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### *Bar Plot Markdown* 
This bar chart breaks the vegetables into 4 types (Canned, Dried, Fresh and Frozen) and shows the average price per one cup size. Fresh and Canned are very close in even prices per cup but frozen is the most expensive form of vegetable to buy on average cup size. The cheapest veggies you can get per cup are the dried kind of veggies. I think it is also low in price becasuse dried is not as popular as the other forms of vetgetables.

```python
import matplotlib.pyplot as plt
import io
```

```python
colors = {'Fresh': 'green', 'Frozen': 'blue', 'Canned': 'purple'}

# Plot each form with a unique color
for form, color in colors.items():
    subset = df[df['Form'] == form]
    plt.scatter(
        subset['RP'], 
        subset['CupEquivalentPrice'], 
        label=form, 
        color=color, 
        s=100, # Marker size
        alpha=0.7
    )

# Add plot labels and title
plt.xlabel('Retail Price $')
plt.ylabel('1 Cup Equivalent Price')
plt.title('Vegetable Forms by Retail Price and Cup Equivalent Price')
plt.legend(title='Form')
plt.grid(True)
plt.show()
```

### *Scatter Plot Markdown*
This scatter plot provides more detail by comparing the average retail price and the average price per cup the vegetable go for. I notice that the Fresh vegetables are more scattered in both retail price and cup equivalent price. Also the greatest cluster of data points for retail price is between $1-$3.
